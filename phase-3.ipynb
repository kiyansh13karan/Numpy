# Advance Operation with Business Examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data Structure : [restaurant_id, 2021, 2022, 2023, 2024]

sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("==== Zomato sales analysis ====")

# shape of the data
print("\n Sales data shape : " , sales_data.shape)

# print the first three restaurant sales data 
print("\n Sample data for first three restaurant : \n" , sales_data[0 : 3]) # slicing

```

### Total sales per year :-

```python
# sum to sales per year of all restaurant (axis=0 , top-to-bottom)
yearly_total = np.sum(sales_data[: , 1 :] , axis=0)
print("\n Sum of all restaurants sales per year : " , yearly_total)

```

### Minimum sales per restaurant :-

```python
# minimum sale kiss saal kitni hui hai hr restaurant ki 
min_sales = np.min(sales_data[: , 1 : ] , axis=1)
print("\n Minimum sales of restaurants : " , min_sales)
```

### Maximum sale of which restaurant per year :-

```python
# sbse jyada sale kiss restaurant ki hui hai ek saal me 
max_sale = np.max(sales_data[:, 1:] , axis=0)
print("\n Maximum sale of restaurant per year : " , max_sale)
```

### Average sales per Restaurant :-

```python
# average sales
avg_sales = np.mean(sales_data[:, 1:] , axis=1) # axis=1 -> left-to-right
print("\n Average sales of restaurant : " , avg_sales)
```

### Cumulative sum of each restaurant :-  

```python
#  the cumulative sum refers to the running total of elements along a given axis in an array
cum_sum = np.cumsum(sales_data[:, 1:] , axis=1)
print("\n Cumulative sum of each restaurant is : \n" , cum_sum)
```

### Graph Plotting :- 

```python
# plotting graph for average of cumulative sum :- 
plt.figure(figsize=(12, 7)) # figsize=(length, breadth)
plt.plot(np.mean(cum_sum, axis=0))
plt.title("Average cumulative sales across all restaurant")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

### Vectors :-

```python
vector1 = np.array([1, 2, 3, 4, 5])
vector2 = np.array([6, 7, 8, 9, 10])

# adding of vectors 
print("\n Addition of vector : " , vector1 + vector2)
# multiplication of vector 
print("\n Multiplication of vector : " , vector1 * vector2)
# dot product in vector
print("\n Dot product of vector : " , np.dot(vector1 , vector2))
# angle of vector 
# angle of vectors = dot product of both vectors / dot product of both vectors magnutide
angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))# for finding magnitude => np.linalg.norm(vector)
print("\n Angle of vectors is : " , angle)
```

### Vectorised Operation :- 

```python
restaurant_types = np.array(['biryani' , 'chinese' , 'pizza' , 'burger' , 'cafe'])

# for converting into upper case 
vectorized_upper = np.vectorize(str.upper)
print("\n Vectorized upper case :- " , vectorized_upper(restaurant_types))

# for converting into lower case 
vectorized_lower = np.vectorize(str.lower)
print("\n Vectorized Lower Case :- " , vectorized_lower(restaurant_types))

# for converting into capitalize case 
vectorized_capitalize = np.vectorize(str.capitalize)
print("\n Vectorized capitalize case :- " , vectorized_capitalize(restaurant_types))
```
